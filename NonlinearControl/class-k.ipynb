Below are examples of class $K$ functions. For reference, a continuous function $\alpha: [0, a) \to [0, \infty)$ is defined as class $K$ if it is strictly increasing and $\alpha(0) = 0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of r values
r = np.linspace(-10, 10, 400)

# Plot (1 - e^{-r})
y = 1 - np.exp(-r)
plt.plot(r, y)
plt.text(2.1, 0.8, r'$\alpha(r) = 1 - e^{-r}$', verticalalignment='center', horizontalalignment='left')

# Plot atan(r)
y = np.arctan(r)
plt.plot(r, y)
plt.text(2.1, 1.2, r'$\alpha(r) = \tan^{-1}(r)$', verticalalignment='center', horizontalalignment='left')

plt.xlabel(r'$r$')
plt.ylabel(r'$\alpha(r)$')
plt.title(r'Example class $\mathcal{K}$ functions')
plt.grid(False)
# plt.legend([r'$\alpha(r) = kr^2$'])

plt.xlim(0, 2)
plt.ylim(0, 1.5)
plt.show()
```

Below are examples of class $K_{\infty}$ functions. For reference, a continuous function $\alpha: [0, \infty) \to [0, \infty)$ is defined as class $K_{\infty}$ if it is class $K$ and $\alpha(r) \to \infty$ as $r \to \infty$

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of r values
r = np.linspace(-10, 10, 400)

# Plot kr^2
y = r**(2)
plt.plot(r, y)
plt.text(2.1, 4.0, r'$\alpha(r) = r^{(3/2)}$', verticalalignment='center', horizontalalignment='left')

# Plot kr 
y = 1.5 * r
plt.plot(r, y)
plt.text(2.1, 3.0, r'$\alpha(r) = 1.5r$', verticalalignment='center', horizontalalignment='left')

# Plot \min{r, r^2}
y = np.minimum(1/2 * r, r**2)
plt.plot(r, y)
plt.text(2.1, 1.0, r'$\alpha(r) = \min(0.5 r, r^2)$', verticalalignment='center', horizontalalignment='left')

plt.xlabel(r'$r$')
plt.ylabel(r'$\alpha(r)$')
plt.title(r'Example class $\mathcal{K}_{\infty}$ functions')
plt.grid(False)
# plt.legend([r'$\alpha(r) = kr^2$'])

plt.xlim(0, 2)
plt.ylim(0, 4)
plt.show()
```

Below are examples of class $KL$ functions. For reference, a continuous function $\beta: [0, a) \times [0,\infty) \to [0, \infty)$ is class $KL$ if, for each fixed $s$, the mapping $\beta(r,s)$ belongs to class $K$ with respect to $r$ and, for each $r$, the mapping $\beta(r,s)$ is decreasing with respect to $s$ and $\beta(r,s) \to 0$ as $s \to \infty$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of t values
s = np.linspace(0, 2, 100)

# Create a meshgrid for r and s values
r = np.linspace(0, 2, 100)
R, S = np.meshgrid(r, s)

# Compute beta(r, s)
alpha = 1
M = 1
beta =  M * np.exp(-alpha * S) * R

# Plot the 3D plot of beta(r, s)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(R, S, beta, cmap='viridis')

ax.set_xlabel('r')
ax.set_ylabel('s')
ax.set_zlabel(r'$\beta(r, s)$')
ax.set_title(r'3D plot of $\beta(r, s) = M e^{-\alpha s} r$ with $M = 1$, $\alpha = 1$')
ax.view_init(elev=30., azim=-65)
fig.set_size_inches(10, 7)
ax.set_box_aspect([1,1,1])  # Aspect ratio is 1:1:1
ax.margins(x=0.1, y=0.1, z=0.1)  # Add padding to the edges

plt.show()
```

```python
# Define the constant k
k = 1

# Compute beta(r, s) for the new function
beta_new = R / (k * S * R + 1)

# Plot the 3D plot of beta(r, s)
fig_new = plt.figure()
ax_new = fig_new.add_subplot(111, projection='3d')
ax_new.plot_surface(R, S, beta_new, cmap='viridis')

ax_new.set_xlabel('r')
ax_new.set_ylabel('s')
ax_new.set_zlabel(r'$\beta(r, s)$')
ax_new.set_title(r'3D plot of $\beta(r, s) = \frac{r}{ksr + 1}$ with $k = 1$')
ax_new.view_init(elev=30., azim=-65)
fig_new.set_size_inches(10, 7)
ax_new.set_box_aspect([1,1,1])  # Aspect ratio is 1:1:1
ax_new.margins(x=0.1, y=0.1, z=0.1)  # Add padding to the edges

plt.show()
```
